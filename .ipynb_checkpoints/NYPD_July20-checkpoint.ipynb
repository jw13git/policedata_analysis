**Data Manipulation**

```python
#Using the function pre_bor() allows us to create a new column named 'borough' which shows which borough
#(Manhatten,Bronx,Brooklyn,Queens,Staten Island) the complaint happened/sunmitted.

def pre_bor(precinct):
    if precinct <= 39:
        return "Manhattan"
    elif precinct >= 40 and precinct < 59:
        return "Bronx"
    elif precinct >=60 and precinct < 99:
        return "Brooklyn"
    elif precinct >= 100 and precinct < 119:
        return "Queens"
    elif precinct >= 120 and precinct < 125:
        return "Staten Island"
    else:
        return "unknown"
pre_bor(50)

nypd['borough'] = np.vectorize(pre_bor)(nypd['precinct'])
```

```python
#verifying the new column is there

nypd.head()
```

```python
#Using the function may() takes in a year and returns the mayor (Guiliani,Bloomberg,DeBlasio) that 
#was in office when the complaint happened/submitted.

def may(mayor):
    if mayor >= 1994 and mayor <= 2001:
        return "Giuliani"
    elif mayor >= 2002 and mayor <= 2013:
        return "Bloomberg"
    elif mayor >= 2014 and mayor <= 2020:
        return "DeBlasio"
    else:
        return "unknown"
may(2012)
```

```python
#Creates a new column labeled "mayor" and runs the function may() through the 'year_received' column and places the result in
#the new column 'mayor'

nypd['mayor'] = np.vectorize(may)(nypd['year_received'])
```

```python
#Checking the column

nypd['mayor'].value_counts(dropna=False)
```

```python
#dropping all unknown mayors aka mayors before Guliani 

nypd = nypd[nypd.mayor != 'unknown']
```

```python
#dropping all null precinct and checking if it was successful

nypd = nypd[nypd['precinct'].notna()]
nypd.info()
nypd.isnull().sum()
```

```python
#First we check what is the youngest complaintant age 
#secondly we drop the rows with a complaintant age below 10 while keeping the nulls

nypd.loc[nypd.complainant_age_incident == nypd.complainant_age_incident.min()]
nypd = nypd[(nypd['complainant_age_incident'] > 10) | (nypd['complainant_age_incident'].isna())]
```

```python
#replacing all 'complaintant_age_incident' nulls with the average complaintant age 

nypd['complainant_age_incident'].fillna((nypd['complainant_age_incident'].mean()),inplace=True)
```

```python
#checking if it was successful

nypd.isnull().sum()
```

```python
#reset the overall index of the table

nypd.reset_index()
```

```python
#Change the 'allegations' strings to lower case and drop all rows that are labeled 'other'  

nypd.allegation = nypd.allegation.str.lower()
nypd = nypd[nypd.allegation.str.contains('other')==False]
```

```python
#Overall checking

nypd.info()
```

```python
#Export current manipulated dataset as a csv

nypd.to_csv("cleaned_nypd.csv",header=True,index=False)
```
